```python
# Discuss string slicing and provide examples

'''
String slicing is a powerful feature in Python that allows you to 
extract a portion (or slice) of a string. Slicing is done 
using the : operator inside square brackets. 

The syntax for string slicing is:
string[start:end:step]

start: The starting index of the slice.
end: The ending index of the slice.
step: The step size or stride between each character in the slice.
'''

#Examples

#1)
a = "Hello, World!"
print(a[0:5])  # Output: Hello

#2)
a = "Hello, World!"
print(a[0:13:2])

#3)
a = "Hello, World!"
print(a[::-1])

#4)
a = "Hello, World!"
print(a[-6:])  # Output: World!

#5)
a = "Hello, World!"
print(a[11:5:-1])

#6)
a = "Hello, World!"
print(a[::3])  # Output: Hl r!
```

```python
#2) Explain the key features of lists in Python

'''
->List are ordered collection of items 
->[] 
->List is like a shopping bag which can store everything
->can store any datatype


'''
#1)Creating a List:
my_list = [1, 2, 3, 4, 5]
print(my_list)  

#2)Indexing and Slicing:
print(my_list[0])  
print(my_list[1:3])   

#3)Appending and Extending:
    
my_list.append(6)
print(my_list)  

my_list.extend([7, 8])
print(my_list)   

#4)Inserting and Removing:

my_list.insert(0, 34)
print(my_list)   

my_list.remove(3)
print(my_list)   

#5)Popping and Clearing:

popped_element = my_list.pop()

print(popped_element)   
print(my_list)  

my_list.clear()
print(my_list)  

#6)Finding Elements:

my_list = [1, 2, 2, 3, 4, 5]
print(my_list.index(2))  
print(my_list.count(2))

#7)Sorting and Reversing:

my_list.sort()
print(my_list) 

my_list.reverse()
print(my_list)  
```

```python
#4) Compare and contrast tuples and lists with examples
'''
Differences

Mutability:
Lists: Mutable, meaning you can modify elements after creation.
Tuples: Immutable, meaning once created, you cannot modify the elements.

Syntax:
Lists: Defined using square brackets [].
Tuples: Defined using parentheses ().

Performance:
Lists: Generally slower due to their mutability.
Tuples: Generally faster due to their immutability, 
making them more memory-efficient.

Use Cases:
Lists: Suitable for collections of items that may need to change over
time (e.g., lists of tasks, collections of user inputs).
Tuples: Suitable for collections of items that should remain
constant throughout the program (e.g., coordinates, fixed configurations).
'''

#1)Creating and Accessing Elements
 
# List
my_list = [1, 2, 3, 4, 5]
print(my_list[0])   

# Tuple
my_tuple = (1, 2, 3, 4, 5)
print(my_tuple[0])   

#2)Modifying Elements

# List - mutable
my_list[1] = 20
print(my_list) 

# Tuple - immutable (will raise an error)
# my_tuple[1] = 20  # TypeError: 'tuple' object does not support 
#item assignment

#3)Adding and Removing Elements

# List - supports adding and removing elements
my_list.append(6)
print(my_list)  

my_list.remove(3)
print(my_list)  

# Tuple - does not support adding or removing elements
# my_tuple.append(6)  # AttributeError: 'tuple' object has no attribute 'append'
# my_tuple.remove(3)  # AttributeError: 'tuple' object has no attribute 'remove'

#4)Indexing and Slicing

# List
print(my_list[1:4])   
# Tuple
print(my_tuple[1:4])  
```

```python
#5) Describe the key features of sets and provide examples of their use


'''
Key Features of Sets

1)Unordered Collection:
Sets do not maintain any order. The elements in a set are stored in an
arbitrary order.

2)Unique Elements:
Sets automatically ensure that all elements are unique. Duplicate 
elements are not allowed.

3)Mutable:
Sets are mutable, meaning you can add and remove elements after the set
has been created.

4)No Indexing or Slicing:
Since sets are unordered, they do not support indexing, slicing, or other 
sequence-like behavior.

5)Efficient Membership Testing:
Sets are optimized for membership tests (checking if an element is in the set) due to their underlying implementation using hash tables.

6)Set Operations:
Sets support standard mathematical operations like union, intersection, 
difference, and symmetric difference.
'''


#1) Removing Duplicates from a List:

my_list = [1, 2, 2, 3, 4, 4, 5]
unique_elements = list(set(my_list))
print(unique_elements)   

#2)Set Operations for Common Elements:

students_class_a = {"Khanjan", "Ajay", "Bijay"}
students_class_b = {"Ajay", "Ayush", "Rahul"}

# Students in both classes
students_class_a & students_class_b

# All students
students_class_a | students_class_b

# Students only in class A
students_class_a ^ students_class_b

#3)Adding and Removing Elements
 
my_set = {1, 2, 3, 4, 5}

# Adding an element
my_set.add(6)
print(my_set)   

# Removing an element
my_set.remove(3)
print(my_set)   

# Discarding an element (removes if present, does nothing if not)
my_set.discard(2)
print(my_set)  

# Popping an element (removes and returns an arbitrary element)
popped_element = my_set.pop()
print(popped_element)   
print(my_set)  

# Clearing all elements
my_set.clear()
print(my_set)  
```

```python
#7) Describe how to add, modify, and delete items on a dictionary with examples

#1)Adding Items to a Dictionary
#You can add items to a dictionary by assigning a value to a new key.

#Example:
my_dict = {"name": "Khanjan", "age": 20}

# Adding a new key-value pair
my_dict["city"] = "Banglore"
print(my_dict)  

#2)Modifying Items in a Dictionary
#To modify an item in a dictionary, you simply assign a new value 
#to an existing key.

#Example:
my_dict = {"name": "Khanjan", "age": 18, "city": "New York"}

# Modifying an existing key-value pair
my_dict["age"] = 24
print(my_dict)   

# Modifying another key-value pair
my_dict["city"] = "Los Angeles"
print(my_dict)   

#3)Deleting Items from a Dictionary
#There are several ways to delete items from a dictionary:

#Using the del Statement:

#Removes the item with the specified key.
 
my_dict = {"name": "Khanjan", "age": 24, "city": "Los Angeles"}

# Deleting an item by key
del my_dict["age"]
print(my_dict) 
```
